# Analysis of Miles per Gallon in Cars Between 1970 and 1982

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
df = sns.load_dataset('mpg')
df
```

## Strong negative correlation exists between mpg and a number of factors, including number of cyliners, displacement, horsepower, and weight of the vehicle

```python
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```

# Horsepower as an example: there is a strong negative correlation with miles per gallon.

```python
sns.lmplot(data=df, x="mpg", y="horsepower")
plt.show()
```

# American cars of the time tended to have higher horsepower, and thus, lower mileage

```python
sns.boxplot(data=df, x="origin", y="mpg")
plt.title("Miles per Gallon by Manufacturing Region")
plt.ylabel("Miles per Gallon")
plt.xlabel("Manufacturing Region")
plt.show()
```

```python
sns.boxplot(data=df, x="origin", y="horsepower")
plt.title("Horsepower by Manufacturing Region")
plt.xlabel("Manufacturing Region")
plt.ylabel("Horsepower")
plt.show()
```
